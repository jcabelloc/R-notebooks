### a) Create a binary variable that takes on a 1 for cars with gas mileage above the median, and a 0 for cars with gas mileage below the median. Use this variable as response in the following analysis.

```r
y = rep(0, length(Auto$mpg))
med_mpg = median(Auto$mpg)
y[Auto$mpg> med_mpg] = 1
y = as.factor(y)
y
```

* By using the same value of cost but different predictors, the number of support vectors change
* By considering all the predictors, we have obtained a best model with a cost of 100. The cross validation error for this setting is 0.084, being the lowest among all the cost values considered

### (c) Now repeat (b), this time using SVMs with radial and polynomial kernels, with different values of gamma, degree and cost. Comment on your results.

### c.1 - SVM radial kernel

```r
xcols = c("horsepower", "weight")
dat = cbind (Auto[xcols], y)
train = sample(length(y), length(y)*0.6)
svmfit = svm(y~., data=dat[train,], kernel="radial", cost=0.1, gamma = 1)
```

```r
plot(svmfit, dat[train,])
```

```r
summary(svmfit)
```

```r
xcols = c("horsepower", "weight")
dat = cbind (Auto[xcols], y)
train = sample(length(y), length(y)*0.6)
svmfit = svm(y~., data=dat[train,], kernel="radial", cost=10000, gamma = 1)
```

```r
plot(svmfit, dat[train,])
```

```r
summary(svmfit)
```

```r
set.seed(1)
xcols = c("cylinders", "displacement", "horsepower", "weight", "acceleration", "year")
dat = cbind (Auto[xcols], y)
gammas = c(0.5,1,2,3,4)
costs = c(0.001, 0.01, 0.1, 1, 10, 100, 1000)
tune.out = tune(svm, y~., data = dat, kernel = "radial", ranges = list(cost=costs, gamma = gammas))
```

```r
summary(tune.out)
```

```r
bestmod = tune.out$best.model
ypred = predict(bestmod, dat)
table(predict=ypred, truth=dat$y)
```

### Comments
* To get intuition, We first obtained two radial models by using only two predictors.
* Then we see that as wee increase the cost, the model tends to overfit the dataset
* The best model is obtained when cost = 1 and gamma = 2. For those values, the cross validation error rate is 0.055
* This best radial SVM model beats the linear support classifier model
